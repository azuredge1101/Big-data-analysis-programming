```python
import csv
import requests
import tkinter as tk
from tkinter import ttk
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.pyplot as plt 
from matplotlib.font_manager import fontManager

def download_and_process_data(year, month):
    visitors = {}
    
matplotlib.rc('font', family='MingLiU')
fontManager.addfont('TaipeiSansTCBeta-Regular.ttf')
matplotlib.rc('font', family='Taipei Sans TC Beta')

def download_and_process_data(year, month):
    visitors = {}
    url = f'http://tcgmetro.blob.core.windows.net/stationod/%E8%87%BA%E5%8C%97%E6%8D%B7%E9%81%8B%E6%AF%8F%E6%97%A5%E5%88%86%E6%99%82%E5%90%84%E7%AB%99OD%E6%B5%81%E9%87%8F%E7%B5%B1%E8%A8%88%E8%B3%87%E6%96%99_{year}{month}.csv'   

    response = requests.get(url)

    if response.status_code == 200:
        with open(f'mule_{year}_{month}.csv', 'wb') as file:
            file.write(response.content)

        with open(f'mule_{year}_{month}.csv', newline='', encoding='utf-8') as file:
            reader = csv.reader(file)
            for row in reader:
                if 'BL板橋' in row:
                    date, visitor = row[0], int(row[4])
                    date = date[8:]
                    visitors[date] = visitors.get(date, 0) + visitor
                    sorted_visitors = dict(sorted(visitors.items()))
        return visitors
    else:
        print('檔案下載失敗')
        return {}

def on_button_click():
    selected_month = month_combobox.get()
    
    visitors_2021 = download_and_process_data('2021', selected_month)
    visitors_2022 = download_and_process_data('2022', selected_month)
    
    dates = list(visitors_2021.keys())
    Visitors_2021 = list(visitors_2021.values())
    Visitors_2022 = list(visitors_2022.values())

    plt.plot(dates, Visitors_2021, marker='o', label='2021')
    plt.plot(dates, Visitors_2022, marker='x', label='2022')
    plt.xlabel('日期')
    plt.ylabel('旅客人數')
    plt.title(f"BL板橋站 {selected_month} 月每日旅客人數比較")
    plt.legend()
    plt.show()

root = tk.Tk()
root.title("2021年與2022年各月份BL板橋站每日旅客人數折線圖查詢系統")

window_width = 500
window_height = 130
root.geometry(f"{window_width}x{window_height}")

label = tk.Label(root, text="請選擇比較的月份")
label.pack(pady=10)

months = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
month_combobox = ttk.Combobox(root, values=months)
month_combobox.set(months[0])
month_combobox.pack(pady=10)

compare_button = tk.Button(root, text="比較", command=on_button_click)
compare_button.pack(pady=10)

root.mainloop()
```
