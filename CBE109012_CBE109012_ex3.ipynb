```python
scores = {}

with open('exam2-input.txt', 'r', encoding='utf-8') as file:
    for line in file:
        n, s = line.strip().split(', ')
        if n in scores:
            scores[n] += int(s)
        else:
            scores[n] = int(s)


names = list(scores.keys())
total_scores = list(scores.values())

print(names)
print(total_scores)
    
import matplotlib
import matplotlib.pyplot as plt 
from matplotlib.font_manager import fontManager
import matplotlib.font_manager as fm
font1 = fm.FontProperties(fname=r'C:\Users\USER\TaipeiSansTCBeta-Regular.ttf')

listx = ['鄭十全','吳二連','金六順','莊四維','張七巧','王一勝','李九州','楊八妹','姚五福','陳三省','吳武德','蔣光彩']
listy = [2237,974,1112,2766,1278,1439,2141,785,936,66,1995,209]
plt.bar(listx, listy, width=0.8)
plt.xticks(rotation=45)
plt.title("積分統計長條圖")
plt.xlabel("姓名")
plt.ylabel("積分")
plt.show()
```
